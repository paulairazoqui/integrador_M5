```python
import pandas as pd
```

```python
listing = pd.read_csv(r'C:\Users\paula\OneDrive\Curso\Henry\M5\Proyecto Integrador\2 AirBnB\listings.csv')

listing.head
```

```python
listing['amenities'].head()
```

```python
listing['amenities'] = listing['amenities'].str.strip('{}')
listing['amenities'].head()
```

```python
listing['amenities'] = listing['amenities'].str.split(',')
listing['amenities'].head()
```

```python
listing['amenities'] = listing['amenities'].apply(lambda x: [amenity.strip('"') for amenity in x])
listing['amenities'].head()
```

```python
# Contar la frecuencia de cada amenity
amenities_counts = listing['amenities'].explode().value_counts()

# Seleccionar las 10 m치s comunes
top_10_amenities = amenities_counts.head(10).index.tolist()
top_10_amenities
```

```python
# Crear columnas binarias para los 10 amenities m치s comunes
for amenity in top_10_amenities:
    listing[amenity] = listing['amenities'].apply(lambda x: 1 if amenity in x else 0)

# Verifica que las columnas se hayan agregado correctamente
listing.head()

```

```python
barrios_bsas = [
    "Agronom칤a", "Almagro", "Balvanera", "Barracas", "Belgrano", "Boedo", "Caballito", "Chacarita",
    "Coghlan", "Colegiales", "Constituci칩n", "Flores", "Floresta", "La Boca", "La Paternal",
    "Liniers", "Mataderos", "Monte Castro", "Monserrat", "Nueva Pompeya", "Nu침ez", "N칰침ez" ,"Palermo",
    "Parque Avellaneda", "Parque Chacabuco", "Parque Chas", "Parque Patricios", "Puerto Madero",
    "Recoleta", "Retiro", "Saavedra", "San Crist칩bal", "San Cristobal", "San Nicol치s","San Nicolas", "San Telmo", "Velez Sarsfield", "V칠lez Sarsfield",
    "V칠lez S치rsfield","Versalles", "Villa Crespo", "Villa del Parque", "Villa Devoto", "Villa Gral. Mitre",
    "Villa Lugano", "Villa Luro", "Villa Ort칰zar", "Villa Pueyrred칩n", "Villa Real", "Villa Riachuelo",
    "Villa Santa Rita", "Villa Soldati", "Villa Urquiza", "Villa General Mitre"
]
```

```python
print(len(barrios_bsas))
```

```python
# Funci칩n para actualizar city y state si el barrio est치 en barrios_bsas
def actualizar_ciudad_estado(row):
    if any(barrio in row['neighbourhood'] for barrio in barrios_bsas):
        row['city'] = "Ciudad Aut칩noma de Buenos Aires"
        row['state'] = "CABA"
    return row

# Aplicar la funci칩n a cada fila
listing = listing.apply(actualizar_ciudad_estado, axis=1)

```

```python
# Remover el s칤mbolo de moneda y las comas, luego convertir a float
listing['price'] = listing['price'].replace('[\$,]', '', regex=True).astype(float)
listing['weekly_price'] = listing['weekly_price'].replace('[\$,]', '', regex=True).astype(float)
listing['monthly_price'] = listing['monthly_price'].replace('[\$,]', '', regex=True).astype(float)
listing['security_deposit'] = listing['security_deposit'].replace('[\$,]', '', regex=True).astype(float)
listing['cleaning_fee'] = listing['cleaning_fee'].replace('[\$,]', '', regex=True).astype(float)
listing['extra_people'] = listing['extra_people'].replace('[\$,]', '', regex=True).astype(float)


print(listing[['id', 'latitude', 'longitude', 'price','extra_people','cleaning_fee','security_deposit','monthly_price','weekly_price']].head())
print(listing[['id', 'latitude', 'longitude', 'price','extra_people','cleaning_fee','security_deposit','monthly_price','weekly_price']].dtypes)
```
